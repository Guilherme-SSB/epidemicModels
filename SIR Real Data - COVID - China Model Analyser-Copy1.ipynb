```python

import requests
import pandas as pd

covid_api = 'https://corona-api.com/countries/'
rest_countries = 'https://restcountries.eu/rest/v2/alpha/'
country = 'CN' # Alpha-2 ISO3166

data_json =  requests.get(covid_api + country).json()
country = requests.get(covid_api + country).json()

N = country['data']['population']

print(country['data']['name'])
```

### Organizando os dados

```python

from datetime import datetime

df = pd.DataFrame(data_json['data']['timeline'])
df = df.sort_values('date').reset_index()

from datetime import datetime, timedelta
df['date'] = [datetime.fromisoformat(f) for f in df['date']]
df = df.drop_duplicates(subset='date', keep = 'last')

# Criando o vetor de tempo
first_date = df['date'].iloc[0]
size_days = (df['date'].iloc[-1] - df['date'].iloc[0]).days
date_vec = [first_date + timedelta(days=k) for k in range(size_days)]

new_df = pd.DataFrame(date_vec, columns=['date'])
new_df = pd.merge(new_df, df, how='left', on= 'date')
new_df = new_df.drop(columns= ['index',  'updated_at', 'is_in_progress'])

for col in new_df.columns[1:]:
    new_df[col] = new_df[col].interpolate(method='polynomial', order=1)
df = new_df.dropna()

df.head()
```

```python

I = df['active'].to_numpy()
R = df['recovered'].to_numpy()
M = df['deaths'].to_numpy()
S = N - R - I

# Creating the time vector
t = np.linspace(0, len(I), len(I))

until = 25

Sd, Id, Md, Rd, td = S[:until], I[:until], M[:until], R[:until], t[:until]
```

```python

import scipy.signal as scs
from scipy.interpolate import Rbf, UnivariateSpline


spl_I = UnivariateSpline(t, I)
spl_R = UnivariateSpline(t, R)
spl_S = UnivariateSpline(t, S)

time = np.linspace(0, len(I), 10*len(I))

I = spl_I(time)
R = spl_R(time)
S = spl_S(time)

until = 40*len(I)
dt = 1 / 40

Sd, Id, Md, Rd, td = S[:until], I[:until], M[:until], R[:until], t[:until]
```

```python

size = len(Id) - 1
Y = np.zeros((3*size, 1))       # Y   = [ 3*n x 1 ]
Phi = np.zeros((3*size, 3))     # Phi = [ 3*n x 2 ]
 
k = 0        # S = [1,2,3,4,5]  e.g. s(k-1) = [1,2,3,4] and s(k) = [2,3,4,5]  
for s, i, r, s_, i_, r_ in zip(Sd[:-1], Id[:-1], Rd[:-1], Sd[1:], Id[1:], Rd[1:]):
    
    # s <- s(k-1)    i <- i(k-1)   r <- r(k-1)
    # s_ <- s(k)     i_ <- i(k)    r_ <- r(k) 
    y_k = np.zeros((3,1))
    y_k[0] = s_            # s(k)   ->  b * (- s(k-1) i(k-1) / N) + s(k-1)
    y_k[1] = i_            # i(k)   ->  b * (s(k-1) i(k-1) / N) - r i(k-1) + i(k-1)
    y_k[2] = r_            # r(k)   ->  r i(k-1) + r(k-1) 
    
    Y[k:k+3,:] = y_k[:,:]     # Y = [ y(0), ..., y(N) ]ÀÜ\top
    
    phi_k = np.zeros((3,3))
    phi_k[0,0] = - dt * s * i / N  # - s(k-1) i(k-1) / N   | - s(k-1) i(k-1) / N,     0    , s(k-1)|
    phi_k[1,0] = dt * s * i / N    # s(k-1) i(k-1) / N     |   s(k-1) i(k-1) / N,  -i(k-1) , i(k-1)|
    phi_k[1,1] = - dt * i          # - i(k-1)              |          0         ,  i(k-1)  , r(k-1)|
    phi_k[2,1] = dt * i            # i(k-1)
    phi_k[0,2] = s
    phi_k[1,2] = i
    phi_k[2,2] = r
                              #                         | b |
    Phi[k:k+3,:] = phi_k[:,:] #                         | r |
                              #                         | 1 |
    k += 3
```

```python

size = len(I) - 1
Phi_sim = np.zeros((3*size, 3))     # Phi = [ 3*n x 2 ]
 
k = 0        # S = [1,2,3,4,5]  e.g. s(k-1) = [1,2,3,4] and s(k) = [2,3,4,5]
for s, i, r, s_, i_, r_ in zip(S[:-1], I[:-1], R[:-1], S[1:], I[1:], R[1:]):
    
    phi_k = np.zeros((3,3))
    phi_k[0,0] = - dt * s * i / N  # - s(k-1) i(k-1) / N     | - s(k-1) i(k-1) / N,     0    |
    phi_k[1,0] = dt * s * i / N    # s(k-1) i(k-1) / N       |   s(k-1) i(k-1) / N,  -i(k-1) |
    phi_k[1,1] = - dt * i          # - i(k-1)                |          0         ,  i(k-1)  |
    phi_k[2,1] = dt * i             # i(k-1)
    phi_k[0,2] = s
    phi_k[1,2] = i
    phi_k[2,2] = r
                                  #                         | b |
    Phi_sim[k:k+3,:] = phi_k[:,:] #                         | r |
    
    k += 3
```

```python

# Y <- Phi * theta
# pi Phi Y <- theta

theta = np.linalg.pinv(Phi) @ Y
```

```python

Y_pred = Phi_sim @ theta

ds_pred = [ ]
di_pred = [ ]
dr_pred = [ ]

ds_sim = [ ]
di_sim = [ ]
dr_sim = [ ]

phi_k = np.zeros((3,3))
phi_k[:,:] = Phi_sim[0:3,:]
for k in range(len(I)-1): 
    # k = [   0  ,   0  ,   0  ,   1  ,   1  ,   1  ,  ]
    # i = [   0  ,   1  ,   2  ,   3  ,   4  ,   5  ,  ]
    # Y = [ ds(0), di(0), dr(0), ds(1), di(1), dr(1),  ]
    ds_pred.append(Y_pred[3*k+0])
    di_pred.append(Y_pred[3*k+1])
    dr_pred.append(Y_pred[3*k+2])
    
    # Focus simulation
    y_k = phi_k @ theta

    ds_sim.append(y_k[0])
    di_sim.append(y_k[1])
    dr_sim.append(y_k[2])
    
    ds = [-dt*y_k[0]*y_k[1]/N,         0.0, y_k[0]]
    di = [ dt*y_k[0]*y_k[1]/N , -dt*y_k[1], y_k[1]]
    dr = [0.0,                   dt*y_k[1], y_k[2]]
    
    for i in range(3):
        phi_k[0,i] = ds[i]
        phi_k[1,i] = di[i]
        phi_k[2,i] = dr[i]

di_pred_2 = N - np.array(ds_pred) - np.array(dr_pred)
di_pred_2 = np.reshape(np.array(di_pred_2), (len(I)-1,))
ds_pred = np.reshape(np.array(ds_pred), (len(I)-1,))
di_pred = np.reshape(np.array(di_pred), (len(I)-1,))
dr_pred = np.reshape(np.array(dr_pred), (len(I)-1,))

ds_sim = np.reshape(np.array(ds_sim), (len(I)-1,))
di_sim = np.reshape(np.array(di_sim), (len(I)-1,))
dr_sim = np.reshape(np.array(dr_sim), (len(I)-1,))
```

# Monte Carlo


Nesta parte, faremos um teste aumentando a quantidade de amostras de treinamento e prevendo o momento do pico da epidemia a medida que mais dias s√£o utilizados para treinamento. Esse estudo vai possibilitar a an√°lise da certeza da previs√£o do pico da epidemia antes desse acontecer.

```python

saved_param = {'r':[], 'beta':[], 'pop':[]}
saved_prediction = []

start_day = 10
pred_t = np.array(range(int(td[-1])))

for i in range(start_day, len(I)):

    dataset = dict(S=Sd[:i], I=Id[:i], R=Rd[:i])
    td_ = td[:i]

    # Create the model
    sir_model = ss.SIR(pop=N, focus=["S", "I", "R"], verbose = False)

    # Adjust the parameters
    sir_model.fit(dataset, td_,
                  search_pop=True,
                  pop_sens=[0.00001,0.05],
                  beta_sens=[100,10000], 
                  r_sens=[100,100])
    
    saved_param['r'].append(sir_model.parameters[0])
    saved_param['beta'].append(sir_model.parameters[1])
    saved_param['pop'].append(sir_model.parameters[2])
    
    saved_prediction.append(sir_model.predict((Sd[0],Id[0], Rd[0]), pred_t))
    
```
